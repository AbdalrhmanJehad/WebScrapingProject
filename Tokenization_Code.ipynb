```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize, PunktSentenceTokenizer, RegexpTokenizer
```

```python
df = pd.read_csv('care_opinion_data.csv',delimiter=';')
```

```python
nltk.download('punkt_tab')
```

```python
texts =df['Story Text'].dropna().tolist()
```

```python
#Sentence Tokenization
sent_tokenized=[]
for text in texts:
  sentences=sent_tokenize(text)
  sent_tokenized.append(sentences)
#print(sent_tokenized)
```

```python
punkt_tokenizer = PunktSentenceTokenizer()
punkt_tokenized = []
for text in texts:
  sentences = punkt_tokenizer.tokenize(text)
  punkt_tokenized.append(sentences)
#print(punkt_tokenized)
```

```python
Pattern = r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<![A-Z]\.)(?<=\.|\?|\!)\s'
regex_tokenizer = RegexpTokenizer(pattern=Pattern, gaps=True)
regex_tokenized = []
for text in texts:
  sentences = regex_tokenizer.tokenize(text)
  regex_tokenized.append(sentences)
```

```python
results = pd.DataFrame({
    'Original Text':texts,
    'sent_tokenized':sent_tokenized,
    'punkt_tokenized':punkt_tokenized,
    'regex_tokenized':regex_tokenized
})
results.to_csv('tokenization_result.csv',index=False)
```

```python
def find_differences(list1, list2):
    differences = []
    for i in range(min(len(list1), len(list2))):
        if list1[i] != list2[i]:
            differences.append(f"Difference at index {i}:\nList 1: {list1[i]}\nList 2: {list2[i]}\n")
    if len(list1) > len(list2):
        for i in range(len(list2), len(list1)):
            differences.append(f"Extra sentence in List 1 at index {i}: {list1[i]}\n")
    elif len(list2) > len(list1):
        for i in range(len(list1), len(list2)):
            differences.append(f"Extra sentence in List 2 at index {i}: {list2[i]}\n")
    return differences

sent_vs_punkt_diff = []
sent_vs_regex_diff = []
punkt_vs_regex_diff = []

for i in range(len(texts)):
    sent_vs_punkt_diff.append(find_differences(sent_tokenized[i], punkt_tokenized[i]))
    sent_vs_regex_diff.append(find_differences(sent_tokenized[i], regex_tokenized[i]))
    punkt_vs_regex_diff.append(find_differences(punkt_tokenized[i], regex_tokenized[i]))

results['sent_vs_punkt_diff'] = sent_vs_punkt_diff
results['sent_vs_regex_diff'] = sent_vs_regex_diff
results['punkt_vs_regex_diff'] = punkt_vs_regex_diff

display(results[['Original Text', 'sent_vs_punkt_diff', 'sent_vs_regex_diff', 'punkt_vs_regex_diff']])
```
